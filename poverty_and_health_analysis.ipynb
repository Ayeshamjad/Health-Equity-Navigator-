```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset from Google Drive or local storage
file_path = 'poverty_dataset.xlsx'  # Update this with your file path
table_1_cleaned = pd.read_excel(file_path, sheet_name='Table_1', skiprows=4)
```

```python
# Clean the dataset
table_1_cleaned.columns = ['Outcome', 'Area_Name', 'Area_Code', 'Population', 'Affected_Total', 'Affected_Percent', 'Notes']
table_1_cleaned = table_1_cleaned.drop(columns=['Notes'])
```

```python
# 1. Identify high poverty areas
high_poverty_areas = table_1_cleaned.sort_values(by='Affected_Percent', ascending=False).head(10)
```

```python

# 2. Calculate correlation between poverty and affected health conditions
poverty_affected_correlation = table_1_cleaned[['Population', 'Affected_Total', 'Affected_Percent']].corr()
```

```python
# 3. Identify regions with high populations but low affected percentage
low_inequality_high_population = table_1_cleaned[table_1_cleaned['Affected_Percent'] < 15].sort_values(by='Population', ascending=False).head(10)
```

Top 10 Areas with the Highest Poverty Rates

```python

# Set a style for the plots
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Affected_Percent', y='Area_Name', data=high_poverty_areas, palette='Reds_r')
plt.title('Top 10 Areas with the Highest Poverty Rates')
plt.xlabel('Poverty Rate (%)')
plt.ylabel('Area Name')
plt.tight_layout()
```

This bar plot shows the 10 regions with the highest poverty rates, with Liverpool and Manchester among the highest at over 40%.

 Correlation heatmap between Population, Affected_Total, and Affected_Percent

```python
plt.figure(figsize=(8, 5))
sns.heatmap(poverty_affected_correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Poverty and Health Conditions')
plt.tight_layout()


plt.show()
```

The heatmap shows a high correlation (0.94) between the number of people affected by health conditions and those living in poverty.

# 3. Plotting regions with high population but low healthcare inequality

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Population', y='Area_Name', data=low_inequality_high_population, palette='Blues_r')
plt.title('Top 10 High Population Areas with Low Healthcare Inequality')
plt.xlabel('Population')
plt.ylabel('Area Name')
plt.tight_layout()
```

This plot highlights areas with large populations but low poverty-related health inequality, such as Leicester and Redbridge.
